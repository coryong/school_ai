### 6.1 다항 회귀

```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
life = pd.read_csv(data_loc + 'life_expectancy.csv')
life.head()

life = life[['Life expectancy', 'Alcohol', 'Percentage expenditure',               
             'Polio', 'BMI', 'GDP', 'Thinness 1-19 years']]
life.dropna(inplace = True)

X = life[['Alcohol', 'Percentage expenditure', 'Polio',
          'BMI', 'GDP', 'Thinness 1-19 years']]
y = life['Life expectancy']

from sklearn.preprocessing import PolynomialFeatures #다항회귀

poly_feature = PolynomialFeatures(degree = 3) #차수 정해주기 3차 다항식
X = poly_feature.fit_transform(X)
```

```python
X_train,X_test,y_train,y_test = train_test_split(X, y, test_size = 0.2)
lin_model = LinearRegression()
lin_model.fit(X_train, y_train)

y_hat_train = lin_model.predict(X_train)
y_hat_test = lin_model.predict(X_test)
plt.scatter(y_train, y_hat_train, color='r')
plt.scatter(y_test, y_hat_test, color='b')
plt.plot([40, 100], [40, 100])
print('Mean squared error:', mean_squared_error(y_test, y_hat_test)) #평균 제곱 오차값이 확연히 줄어드는 것을 볼 수 있음
```

### 6.3 과적합과 과소적합 - 공짜 점심은 없다

- 과적합 : 학습에 사용된 데이터에 대해서는 좋은 성능을 보이지만, 새로운 데이터에 대해 적용하면 성능이 잘 나오지 않는, 즉 일반화 능력이 떨어지는 모델

- 과소적합: 학습데이터를 설명하는 함수를 제대로 찾지 못하여 학습 데이터와 다른 데이터 모두에 대해 예측을 제대로 하지 못하는 모델을 만드는 일 

### LAB 6-1: 다항 회귀의 회귀 함수를 그려 보자

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.linear_model import LinearRegression

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc+'nonlinear.csv')

plt.scatter(df['x'], df['y']) #데이터를 그려봄 
```

```python
from sklearn.preprocessing import PolynomialFeatures
X = df['x'].to_numpy()
y = df['y'].to_numpy()
X = X.reshape(-1,1)     # 입력을 2차원 벡터가 되게 한다. shape = (m, 1)
feature_cubic = PolynomialFeatures(degree = 3) #3차 다항 회귀
X_3 = feature_cubic.fit_transform(X) # 3차 다항 회귀 입력 데이터 계산
```

```python
lin_model = LinearRegression() # 선형 회귀 모델 객체 생성
domain = np.linspace(0, 1, 100).reshape(-1,1) # 입력은 2차원 벡터로 변형, 0에서 1까지 100개의 구간으로 구성된 입력 벡터 
```

```python
lin_model.fit(X_3, y) #X_3과 y 학습 
domain_3 = feature_cubic.fit_transform(domain) # 3차 다항 회귀 입력 데이터 계산
predictions = lin_model.predict(domain_3) # 예측값 계산
plt.scatter(df['x'], df['y'])
plt.scatter(domain, predictions, color='r')
```

### 6.12 사이킷런을 이용한 서포트 벡터 머신 사용하기

```python
import pandas as pd
import numpy as np

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'two_classes.csv')
df.tail(5)
```

```python
df_positive = df[df['y']>0]     # y가 1인 데이터만 추출
df_negative = df[df['y']==0]    # y가 0인 데이터만 추출
import matplotlib.pyplot as plt
plt.scatter(df_positive['x1'], df_positive['x2'], color='r')
plt.scatter(df_negative['x1'], df_negative['x2'], color='g')
```

```python
from sklearn.svm import LinearSVC
X = df[['x1', 'x2']].to_numpy()           # x1, x2를 입력 벡터로 한다
y = df['y']                               # y열의 값이 레이블 
svm_simple = LinearSVC(C=1, loss='hinge') # SVM 클래스 생성, C:슬랙 변수 최적화의 가중치
svm_simple.fit(X, y)                      # 입력과 레이블로 SVM 학습 실시
```

```python
svm_simple.predict([[0.12, 0.56], [-4, 40],  [0, 40], [5,20]]) #4개의 데이터를 넣고 예측값 보기
```

```python
from yellowbrick.contrib.classifier import DecisionViz
viz = DecisionViz(svm_simple, title="linear SVM") #시각화
viz.fit(X, y)
viz.draw(X, y)
```

### 6.13 파이프라인을 이용한 데이터 정제

```python
# 정규화 표준화를 쓸수도 있지만 사이킷런 패키지의 파이프라인 사용으로 편리하게
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

svm_std = Pipeline([                       # 파이프라인으로 SVM 객체 구성
  ("std", StandardScaler()),               # 데이터 표준화 단계 포함
  ("lsvm", LinearSVC(C=1, loss='hinge')),  # 선형 SVM 분류기 포함
])
svm_std.fit(X, y)
```

```python
from yellowbrick.contrib.classifier import DecisionViz
viz = DecisionViz(svm_std, title="Standardization - SVM")
viz.fit(X, y)
viz.draw(X, y)
```

### 6.14 다항 특정 변환을 통한 비선형 서포트 벡터 머신의 구현

```python
from sklearn.preprocessing import PolynomialFeatures
svm_poly = Pipeline([
  ("std", StandardScaler()), #입력 데이터 표준화
  ("poly_inputs", PolynomialFeatures(degree=5)), #다항 특징 변환 단계 (차수:5)
  ("lsmv", LinearSVC(C=0.01, loss="hinge"))
])
svm_poly.fit(X, y)
viz = DecisionViz(svm_poly, title="polynomial feature SVM")
viz.fit(X, y)
viz.draw(X, y) #직선이 아닌 비선형 경계로 나타냄 
```

### LAB 6-4: 비선형 SVM을 이용한 데이터 분류

```python
import pandas as pd
import numpy as np

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'twisted_data.csv')
print(df.tail(5))
```

```python
#선형 분리가 불가능한 상태로 섞여있음 
df_positive = df[df['y']>0]
df_negative = df[df['y']==0]
import matplotlib.pyplot as plt
plt.scatter(df_positive['x1'], df_positive['x2'], color='r')
plt.scatter(df_negative['x1'], df_negative['x2'], color='g')
```

```python
from sklearn.svm import LinearSVC
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.preprocessing import PolynomialFeatures
from yellowbrick.contrib.classifier import DecisionViz
X = df[['x1', 'x2']].to_numpy()
y = df['y']
```

```python
polynomial_svm_clf = Pipeline([
   ("scaler", StandardScaler()),
   ("poly_features", PolynomialFeatures(degree=5)),
   ("svm_clf", LinearSVC(C=1, loss="hinge"))
])
polynomial_svm_clf.fit(X, y)
viz = DecisionViz(polynomial_svm_clf, title="polynomial feature SVM")
viz.fit(X, y)
viz.draw(X, y)
```

### 6.15 커널 트릭을 이용한 비선형 서포트 벡터 머신

- 다항 특징 변환 사용할 경우 차수가 높아지면 입력되는 특징 벡터의 차원이 급격히 증가 학습효율 낮아짐
- 위 문제 피하고자 커널트릭
- 특징 벡터의 차원을 늘릴 필요 x -> 원래의 입력들 사이의 내적을 구한 뒤에 그 결과만 제곱해도 동일한 결과 

### LAB 6-5: 커널 트릭을 이용한 비선형 SVM

```python
import pandas as pd
import numpy as np

data_loc = 'https://github.com/dknife/ML/raw/main/data/'
df = pd.read_csv(data_loc + 'twisted_data.csv')

from sklearn.svm import SVC  # 커널 트릭을 사용할 수 있는 SVM 분류기
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler
from yellowbrick.contrib.classifier import DecisionViz

X = df[['x1', 'x2']].to_numpy()
y = df['y']
```

```python
polynomial_svm_clf = Pipeline([
   ("scaler", StandardScaler()),
   ("svm_clf", SVC(C=10, kernel = 'poly', degree = 4, coef0=2)) #커널트릭이 가능한 SVC로 생성, 다항 커널의 경우 poly를 커널 키워드 매개변수에 지정, coef0는 커널에 있는 독립항 r
]) #degree와 coef0 파라미터를 변경해서 다양한 결과를 볼 수 있다.
polynomial_svm_clf.fit(X, y)
viz = DecisionViz(polynomial_svm_clf, title="polynomial feature SVM")
viz.fit(X, y)
viz.draw(X, y)
```
